```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
ndata = pd.read_csv("NextData.csv")
```

```python
ndata.columns
```

```python
ndata = ndata.drop(['Advertiser ID', 'Advertiser Status',
       'Advertiser Integration Code', ], axis=1)
```

```python
ndata.columns
```

```python
ndata.columns=['Advertiser', 'TOD', 'Country', 'DeviceType',
       'Currency', 'Impressions',
       'MeasurableImpressions',
       'ViewableImpressions', 'Clicks',
       'Spend']
```

```python
ndata.head()
```

```python
ndata["CTR"] = ndata.Clicks / ndata.Impressions
ndata["Viewability"] = ndata.ViewableImpressions / ndata.MeasurableImpressions
ndata["CPM"] = ndata.Spend / ndata.Impressions * 1000
```

```python
ndata.head()
```

```python
ndata.tail(15)
```

```python
ndata = ndata.drop(ndata.index[336:])
```

```python
ndata.tail()
```

```python
#below "ViewGroup" needs to be established first so the column label is created then the filtering
# & needs to be used 
#syntax is ndata["ViewGroup"] first then the [filtering]
ndata["ViewGroup"] = 0
ndata["ViewGroup"][ndata.Viewability < 0.29999] = 0.3
ndata["ViewGroup"][(0.3 < ndata.Viewability) & (ndata.Viewability < 0.39999)] = 0.4
ndata["ViewGroup"][(0.4 < ndata.Viewability) & (ndata.Viewability < 0.49999)] = 0.5
ndata["ViewGroup"][(0.5 < ndata.Viewability) & (ndata.Viewability < 0.59999)] = 0.6
ndata["ViewGroup"][(0.6 < ndata.Viewability) & (ndata.Viewability < 0.69999)] = 0.7
ndata["ViewGroup"][(0.7 < ndata.Viewability) & (ndata.Viewability < 0.79999)] = 0.8
ndata["ViewGroup"][ndata.Viewability > 0.79999] = 0.9
```

```python
ndata["CTRGroup"] = 0
ndata["CTRGroup"][ndata.CTR < 0.0005] = "Low"
ndata["CTRGroup"][(ndata.CTR > 0.0005) & (ndata.CTR < 0.001)] = "Medium"
ndata["CTRGroup"][ndata.CTR > 0.001] = "High"
```
