```python
import pandas as pd
import numpy as np
```

```python
ndata = pd.read_csv("NextData.csv")
```

```python
ndata.head()
```

```python
ndata = ndata.drop(ndata.index[336::])
```

```python
ndata = ndata.drop(["Advertiser ID", "Advertiser Status", "Advertiser Integration Code"], axis=1)
```

```python
ndata.tail()
```

```python
ndata.columns
```

```python
ndata.columns=['Advertiser', 'TOD', 'Country', 'DeviceType',
       'AdvertiserCurrency', 'Impressions',
       'MeasurableImpressions',
       'ViewableImpressions', 'Clicks',
       'Spend']
```

```python
ndata.head()
```

```python
ndata["CTR"] = ndata.Clicks / ndata.Impressions
```

```python
ndata["CPM"] = ndata.Spend / ndata.Impressions * 1000
```

```python
ndata.Country = ndata.Country.astype("category")
ndata.DeviceType = ndata.DeviceType.astype("category")
```

```python
ndata.head()
```

---

```python
#to do pivot table we have to use pd.pivot_table() rather than pivot()
#syntax is pivot_table(index="", columns="", values=""), index is the row
ndata.pivot_table(index="DeviceType", columns="Country", values="Clicks")
```

```python
#below is 2 levels of rows, basically just pass multiple dimensions into index
ndata.pivot_table(index=["AdvertiserCurrency", "DeviceType"], columns="TOD", values="Clicks")
```

```python
#on above we can see a lot of NaN is applied when the data is not applicable in some case
#to set them to 0 we use fill_value=integer feature
ndata.pivot_table(index="DeviceType", columns="Country", values="Clicks", fill_value=0)
```

```python
#by default when pivoting it's set to return MEAN value 
#here we use aggfunc= to manipulate it
#it can take a single value but also multple value
#we use numpy feature to pass it the command
ndata.pivot_table(index="DeviceType", columns="Country", values="Clicks", aggfunc=np.sum, fill_value=0)
```

```python
#continuing above below example passes multple functions from np to aggfunc
#np.sum will sum all the clicks and max return the largest value and len to count
ndata.pivot_table(index="DeviceType", columns="Country", values="Clicks", aggfunc=[np.sum, np.max, len], fill_value=0)
```

```python
#here continuing above but pass specific function to specific values using {}
ndata.pivot_table(index="Country", columns="DeviceType", values=["Impressions", "Clicks"],
                  aggfunc={"Impressions":np.sum, "Clicks":np.mean}, fill_value=0)
```

```python
#and to add the total value at the bottom we use margins=True
ndata.pivot_table(index=["Country", "DeviceType"], values=["Impressions", "Clicks"],
                  aggfunc={"Impressions":np.sum, "Clicks":np.mean}, fill_value=0, margins=True)
```
