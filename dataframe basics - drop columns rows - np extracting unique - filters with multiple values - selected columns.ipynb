```python
import pandas as pd
```

```python
import numpy as np
```

```python
ndata = pd.read_csv("NextData.csv")
```

```python
ndata.head()
```

```python
ndata.tail(20)
```

```python
#below removes rows, rows from 336 onward
#it uses index to represent rows
ndata = ndata.drop(ndata.index[336:])
```

```python
ndata.tail(10)
```

```python
#below removes rows based upon certain criteria
#below say removing rows with impressions less than 1000
ndata = ndata.drop(ndata.index[ndata.Impressions < 10000])
```

```python
ndata.Impressions.min()
```

```python
#above shows that the syntax works on removing all rows with impressions less than 10000
```

```python
#below removes columns
#axis=1 is used to denote to have columns removed rather than rows
ndata = ndata.drop(["Advertiser ID", "Advertiser Status", "Advertiser Integration Code"], axis=1)
```

```python
ndata.head()
```

```python
ndata.columns=["Advertiser", "TOD", "Country", "DeviceType", "AdvertiserCurrency", "Impressions", "MeasurableImpressions", "ViewableImpressions", "Clicks", "Spend"]
```

```python
ndata.head()
```

```python
ndata["CTR"] = ndata.Clicks / ndata.Impressions
```

```python
ndata.head()
```

```python
ndata["Viewability"] = ndata.ViewableImpressions / ndata.MeasurableImpressions
```

```python
ndata.head(10)
```

```python
ndata.info()
```

```python
ndata["AdvertiserCurrency"] = ndata["AdvertiserCurrency"].astype("category")
ndata["Country"] = ndata["Country"].astype("category")
ndata["DeviceType"] = ndata["DeviceType"].astype("category")
```
